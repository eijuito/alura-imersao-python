```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("acoes_pura.xlsx", sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("acoes_pura.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("acoes_pura.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("acoes_pura.xlsx", sheet_name="ChatGPT")
df_chatgpt
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left').copy()
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtd_teorica'}).copy()
df_principal
```

```python
df_principal['resultado'] = df_principal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left').copy()
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da Empresa', how='left').copy()
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal = df_principal.rename(columns={'Idade (anos)':'Idade'}).copy()
df_principal
```

```python
df_principal['cat_idade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal
```

```python
maior = df_principal['variacao_rs'].max()
menor = df_principal['variacao_rs'].min()
media = df_principal['variacao_rs'].mean()
media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['variacao_rs'].mean()
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['variacao_rs'].mean()
# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Media\tR$ {media:,.2f}")
print(f"Media de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Media de quem desceu\tR$ {media_desceu:,.2f}")

```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('resultado')['variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
# Formatar os valores da coluna 'variacao_rs' em reais
df_analise_saldo['variacao_rs_formatado'] = df_analise_saldo['variacao_rs'].map('R${:,.2f}'.format)
fig = px.bar(df_analise_saldo, x='resultado', y='variacao_rs', text='variacao_rs_formatado', title='Variação Reais por Resultado')
fig.show()
```

```python
fig = px.pie(df_analise_segmento, names='Segmento', values='variacao_rs', labels='variacao_rs', title='Variação Reais por Segmento')
fig.show()
```

```python
df_analise_faixa_etaria = df_principal_subiu.groupby('cat_idade')['variacao_rs'].sum().reset_index()
df_analise_faixa_etaria['variacao_rs'] = df_analise_faixa_etaria['variacao_rs']
df_analise_faixa_etaria

```

```python
# Formatar os valores da coluna 'variacao_rs' em reais
df_analise_faixa_etaria['variacao_rs_formatado'] = df_analise_faixa_etaria['variacao_rs'].map('R${:,.2f}'.format)
fig = px.bar(df_analise_faixa_etaria, x='cat_idade', y='variacao_rs', text='variacao_rs_formatado', title='Variação Reais por Categoria de Idade')
fig.show()
```
